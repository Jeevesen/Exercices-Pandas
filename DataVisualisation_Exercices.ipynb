## Visualisation dâ€™un nuage de points et exploration des variables


```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('house_pricing.csv')
sns.regplot(x=df["LotArea"], y=df["SalePrice"],
            fit_reg=False)
plt.xlim(20000)
plt.ylim(500000)
plt.show()

sns.regplot(x=df["LotFrontage"], y=df["LotArea"],
            fit_reg=False)
```
