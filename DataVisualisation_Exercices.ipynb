## Visualisation dâ€™un nuage de points et exploration des variables


```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

hp = pd.read_csv('./data/house_pricing.csv')
hp.info()
```

```python
sns.set_theme()

sns.scatterplot(x="LotArea", y="SalePrice",data=hp[(hp["LotArea"]<20000)&(hp["SalePrice"]<500000)])
# plt.xlim(0,20000)
# plt.ylim(0,500000)
plt.show()
```

```python
sns.scatterplot(x="LotFrontage", y="LotArea",data=hp[(hp["LotFrontage"]<200)&(hp["LotArea"]<100000)])
# plt.xlim(0,200)
# plt.ylim(0,100000)
plt.show()
```

## Visualisation de relations continues

```python
sp = pd.read_csv('./data/sales_predictions.csv')
sp.sample(50)
```

```python
sp['date'] = pd.to_datetime(sp['date'])
sns.relplot(x="date", y="item_price",data=sp,kind="line").fig.set_size_inches(20,6)
# plt.figure(figsize=(20,6))
plt.show()
```

## Visualisation de variables catÃ©goriques

```python
ibm = pd.read_csv('./data/IBM_HR_ATTRITION.csv')
sns.catplot(x="Attrition",y="JobSatisfaction",data=ibm)
```

```python
sns.boxplot(x="Attrition",y="JobSatisfaction",data=ibm)
```

```python
sns.boxplot(x="Attrition",y="TotalWorkingYears",data=ibm)
```

```python
sns.boxplot(x="Attrition",y="TotalWorkingYears",data=ibm[(ibm["TotalWorkingYears"]<20)])
```
