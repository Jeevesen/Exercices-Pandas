## Visualisation dâ€™un nuage de points et exploration des variables


```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

hp = pd.read_csv('./data/house_pricing.csv')
hp.info()
```

```python
sns.set_theme()

sns.scatterplot(x="LotArea", y="SalePrice",data=hp[(hp["LotArea"]<20000)&(hp["SalePrice"]<500000)])
# plt.xlim(0,20000)
# plt.ylim(0,500000)
plt.show()
```

```python
sns.scatterplot(x="LotFrontage", y="LotArea",data=hp[(hp["LotFrontage"]<200)&(hp["LotArea"]<100000)])
# plt.xlim(0,200)
# plt.ylim(0,100000)
plt.show()
```
